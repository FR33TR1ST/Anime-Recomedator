## Adding the thriller DataFrame to a general DF

```
columns = ['name', 'category','category2', 'category3', 'category4', 'category5', 'category6', 'category7', 'category8', 'category9', 'watch']
general_df = pd.DataFrame(columns=columns)
df = thriller_df
```

```
df.insert(3, 'category2', "-")
df.insert(4, 'category3', "-")
df.insert(5, 'category4', "-")
df.insert(6, 'category5', "-")
df.insert(7, 'category6', "-")
df.insert(8, 'category7', "-")
df.insert(9, 'category8', "-")
df.insert(10,'category9', "-")
```

```
general_df.set_index('name', inplace=True)
```

```
num = 0
num1 = len(df)
for i in range(0, num1):
    objeto = df.iloc[num]
    num+=1
    if objeto[0] not in df.index:
        serie = pd.Series({  "category":objeto[1],
                             "chapters":objeto[2],
                             "category2":objeto[3],
                             "category3":objeto[4],
                             "category4":objeto[5],
                             "category5":objeto[6],
                             "category6":objeto[7],
                             "category7":objeto[8],
                             "category8":objeto[9],
                             "category9":objeto[10],
                             "watch":objeto[11]}, name=objeto[0])
        general_df = general_df.append(serie)
        #category 2
    elif general_df.loc[objeto[0]][2] == '-':
        general_df['category2'][objeto[0]] = objeto[1]    
    #Category 3
    elif general_df.loc[objeto[0]][3] == '-':
        general_df['category3'][objeto[0]] = objeto[1]
    #category 4
    elif general_df.loc[objeto[0]][4] == '-':
        general_df['category4'][objeto[0]] = objeto[1]
    #category 5
    elif general_df.loc[objeto[0]][5] == '-':
        general_df['category5'][objeto[0]] = objeto[1]        
    #category 6
    elif general_df.loc[objeto[0]][6] == '-':
        general_df['category6'][objeto[0]] = objeto[1]        
    #category 7
    elif general_df.loc[objeto[0]][7] == '-':
        general_df['category7'][objeto[0]] = objeto[1]        
    #category 8
    elif general_df.loc[objeto[0]][8] == '-':
        general_df['category8'][objeto[0]] = objeto[1]        
    #category 9
    elif general_df.loc[objeto[0]][9] == '-':
        general_df['category9'][objeto[0]] = objeto[1]          
```
